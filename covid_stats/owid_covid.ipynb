```python
import pandas as pd
import pylab as plt
import numpy as np
import seaborn as sns
import matplotlib as mpl
import pylab as plt
import matplotlib.dates as mdates
sns.set_context('talk')
pd.set_option('display.width', 150)
locator = mdates.AutoDateLocator(minticks=4, maxticks=10)
formatter = mdates.ConciseDateFormatter(locator)
```

```python
df=pd.read_csv('owid-covid-data.csv')
df['date']=pd.to_datetime(df.date)
```

```python
df.columns
```

```python
p=pd.pivot_table(df,index='date',columns='location',values='new_cases_smoothed_per_million')
```

```python
countries=['Ireland','Sweden','United Kingdom']
x=p[countries]
f,ax=plt.subplots(1,1,figsize=(18,5))
x.plot(ax=ax)
sns.despine()
```

```python
sub=df[df.location=='Ireland']
f,ax=plt.subplots(2,1,figsize=(18,10))
axs=ax.flat
sub.plot(x='date',y=['new_cases_smoothed_per_million'],ax=axs[0])
sub.plot(x='date',y=['new_vaccinations_smoothed_per_million'],ax=axs[1])
sns.despine()
```
