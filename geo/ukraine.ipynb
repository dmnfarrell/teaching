```python
import numpy as np
import pandas as pd
import geopandas as gpd
import pylab as plt

import contextily as cx
import folium
```

```python
#gdf = gpd.read_file('ukr_adm_sspe_20220131/ukr_admbnda_adm0_sspe_20220114.shp')
gdf = gpd.read_file('ukr_adm_sspe_20220131/ukr_admbnda_adm1_sspe_20220114.shp')
gdf = gdf.to_crs(epsg=3857)
```

```python
ax = gdf.plot(alpha=0.5, fc='none', ec='r',figsize=(22, 14))
ax.set_axis_off()
cx.add_basemap(ax, source=cx.providers.Stamen.TonerLite)
```

```python
ax = gdf.plot(column='ADM1_EN',alpha=0.5, ec='k',figsize=(22, 14))
ax.set_axis_off()

cx.add_basemap(ax, source=cx.providers.Stamen.TerrainBackground)
cx.add_basemap(ax, source=cx.providers.Stamen.TonerLabels)
```
